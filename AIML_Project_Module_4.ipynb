```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.01, precision=0.0001, max_iter=1000):
    """
    Performs gradient descent to find a local minimum.

    Parameters:
        derivative_func: The derivative function of the target function.
        initial_guess: Initial value of x to start the descent.
        multiplier: Step size (learning rate).
        precision: Precision for stopping the descent.
        max_iter: Maximum number of iterations to prevent infinite loops.

    Returns:
        local_min: The x-value of the local minimum.
        list_x: List of x-values during the descent.
        deriv_list: List of derivative values during the descent.
    """
    x = initial_guess
    list_x = [x]
    deriv_list = [derivative_func(x)]

    for _ in range(max_iter):
        gradient = derivative_func(x)
        new_x = x - multiplier * gradient  # Update x using the gradient
        step_size = abs(new_x - x)

        list_x.append(new_x)
        deriv_list.append(derivative_func(new_x))

        if step_size < precision:
            break

        x = new_x

        # Prevent runaway values
        if abs(x) > 1e6:  # Limit x to a reasonable range
            raise OverflowError("x value exceeded the allowed range during gradient descent.")

    return x, list_x, deriv_list
```

```python
## TASK-4 : Code the same things as they were done for the first two examples

import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x) and its derivative dg(x)
def g(x):
    return x**3 - 3*x**2 + 2  # Example cubic function

def dg(x):
    return 3*x**2 - 6*x  # Derivative of g(x)


```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.001, precision=0.0001, max_iter=1000):
    """
    Performs gradient descent to find a local minimum.

    Parameters:
        derivative_func: The derivative function of the target function.
        initial_guess: Initial value of x to start the descent.
        multiplier: Step size (learning rate).
        precision: Precision for stopping the descent.
        max_iter: Maximum number of iterations to prevent infinite loops.

    Returns:
        local_min: The x-value of the local minimum.
        list_x: List of x-values during the descent.
        deriv_list: List of derivative values during the descent.
    """
    x = initial_guess
    list_x = [x]
    deriv_list = [derivative_func(x)]

    for iteration in range(max_iter):
        gradient = derivative_func(x)
        new_x = x - multiplier * gradient  # Update x using the gradient
        step_size = abs(new_x - x)

        list_x.append(new_x)
        deriv_list.append(derivative_func(new_x))

        if step_size < precision:
            print(f"Converged in {iteration} iterations.")
            break

        x = new_x

        # Prevent runaway values
        if abs(x) > 1e3:  # Adjust the range if needed
            raise OverflowError("x value exceeded the allowed range during gradient descent.")

    return x, list_x, deriv_list
```

```python
## TASK-8 : Code the same things as they were done for the first two examples

import numpy as np

# Define the first function f(x, y)
def f1(x, y):
    return 13 - x**2 - y**2 + 1

# Define the gradient (partial derivatives) of f1(x, y)
def df1(x, y):
    df_dx = -2 * x
    df_dy = -2 * y
    return np.array([df_dx, df_dy])

# Define the second function f(x, y)
def f2(x, y):
    r = 3 - x**2 - y**2
    return 1 / (r + 1)

# Define the gradient (partial derivatives) of f2(x, y)
def df2(x, y):
    r = 3 - x**2 - y**2
    df_dx = -2 * x / ((r + 1)**2)
    df_dy = -2 * y / ((r + 1)**2)
    return np.array([df_dx, df_dy])


```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.1, precision=0.0001, max_iter=100):
    x, y = initial_guess
    history = [(x, y, f1(x, y))]  # Store initial guess and function value

    for _ in range(max_iter):
        grad = derivative_func(x, y)
        x -= multiplier * grad[0]
        y -= multiplier * grad[1]

        history.append((x, y, f1(x, y)))  # Append new values to history

        # Stop if the change is small enough (precision threshold)
        if np.linalg.norm(grad) < precision:
            break

    return (x, y), history
```

## Batch Gradient Descent with SymPy

```python
# Setup
import numpy as np
import sympy as sp

# Define the symbols
a, b = sp.symbols('a b')

# Define the function
f = a**2 + b**2  # Example function

# Define the multiplier and max iterations
multiplier = 0.1
max_iter = 500

# Initial guess
params = np.array([1.8, 1.0])

# Gradient Descent Loop
for n in range(max_iter):
    # Compute the gradients (partial derivatives)
    gradient_x = sp.diff(f, a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = sp.diff(f, b).evalf(subs={a: params[0], b: params[1]})

    gradients = np.array([gradient_x, gradient_y])

    # Update the parameters
    params = params - multiplier * gradients

# Results
print('Values in gradient array', gradients)
print('Minimum occurs at x value of: ', params[0])
print('Minimum occurs at y value of: ', params[1])
print('The cost is: ', f.subs({a: params[0], b: params[1]}))
```

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$

### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
